```
# import the library
import folium
import pandas as pd
```

```
# Sample Project
# https://python-graph-gallery.com/313-bubble-map-with-folium/
# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lat':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5],
   'lon':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97],
   'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador'],
   'value':[10,12,40,70,23,43,100,43]
})
data
```

```
# Make an empty map
m = folium.Map(location=[abs(data['lon'].min()),abs(data['lon'].max())], tiles="OpenStreetMap", zoom_start=2)
```

```
# I can add marker one by one on the map
for i in range(0,len(data)):
   folium.Circle(
      location=[data.iloc[i]['lon'], data.iloc[i]['lat']],
      popup=data.iloc[i]['name'],
      radius=int(data.iloc[i]['value']*10000),
      color='crimson',
      fill=True,
      fill_color='crimson'
  ).add_to(m)
```

```
# Durham Meals Data
dm = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRvNANld7G8-GL2OKB8DdTYHkxlaCz6RNyXIYypRKiaoqecIbHyuxAEKeiM5Bzr4fgSao-7290O87r-/pub?gid=1503161031&single=true&output=csv')
```

```
dm['average'] = (dm['Breakfast'] + dm['Lunch'])/2
dm['max'] = dm[['Breakfast','Lunch']].max(axis=1)

dm.head(1)
# dm[['Site Name','Latitude','Longitude','Breakfast','Lunch','max','average']].head()
```

```
m = folium.Map(location=[dm['Latitude'].mean(),dm['Longitude'].mean()], tiles=None, zoom_start=12)
folium.TileLayer (tiles='openstreetmap', show=True, control=False).add_to(m)
```

```
# I can add marker one by one on the map
multiplier = 0.2
bf = folium.FeatureGroup(name='Breakfast Sites', show=True)
lunch = folium.FeatureGroup(name='Lunch Sites', show=True)

for i in range(0,len(dm)):
  folium.Circle(
    location=[dm.iloc[i]['Latitude'], dm.iloc[i]['Longitude']],
    popup= str(dm.iloc[i]['Site Name']) + "\n Breakfast: " + str(dm.iloc[i]['Breakfast']),
    radius=float(dm.iloc[i]['Breakfast'])*multiplier,
    color='crimson',
    stroke=False,
    fill=True,
    fill_color='crimson',
    control=True,
  ).add_to(bf)
  
  folium.Circle(
    location=[dm.iloc[i]['Latitude'], dm.iloc[i]['Longitude']],
    popup= str(dm.iloc[i]['Site Name']) + "\n Lunch: " + str(dm.iloc[i]['Lunch']),
    radius=float(dm.iloc[i]['Lunch'])*multiplier,
    color='blue',
    stroke=False,
    fill=True,
    fill_color='blue',
    control=True,
  ).add_to(lunch)

bf.add_to(m)
lunch.add_to(m)

folium.LayerControl(collapsed=False).add_to(m)
```
